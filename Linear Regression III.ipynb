```python
import pandas as pd
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

1. Önceki derste olduğu gibi, hedef değişkeninizin sıcaklık ve görünür sıcaklık arasındaki fark olduğu bir doğrusal regresyon modeli oluşturun. Açıklayıcı değişkenler olarak nem ve rüzgar hızı kullanın. Şimdi, modelinizi OLS kullanarak tahmin edin. R-kare ve ayarlanmış R-kare değerleri nelerdir? Tatminkar olduklarını düşünüyor musunuz? Açıklayın.

```python
X = weather_df[["Humidity","Wind Speed (km/h)"]]
Y = weather_df["Temperature (C)"] - weather_df["Apparent Temperature (C)"]
lrm = LinearRegression()
lrm.fit(X, Y)

x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- The r squared and adjusted r squared values are 0.288 for two of them. These values are quite low. It will be good for us if it is at least over 0.70. Too much may mean overfitting.

2. Daha sonra, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Şimdi, bu modelin R-kare değeri nedir? Bu model bir öncekine göre gelişti mi?

```python
df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
df.head()
```

1. Ev fiyatları modelinizi tekrar çalıştırın ve F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC kullanarak modelinizin uygunluğunu değerlendirin.

```python
dummies = pd.get_dummies(df[["SaleType","Street"]],drop_first=True)
X = pd.concat([df["FullBath"],dummies],axis=1,ignore_index=False)
Y = df["SalePrice"]

lrm = LinearRegression()
lrm.fit(X, Y)

x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- The model's R squared, adjusted R squared, and F test values ​​are very low, which is not what we want. For AIC and BIC, the less the better, but we see that the values ​​are very high.

2. Modelinizin tatmin edici olduğunu düşünüyor musunuz? Açıklayın

- Our model is certainly not satisfactory. As I said in the previous question, the values ​​are not at the level we want.

3. Modelinizin uygunluğunu iyileştirmek için, bazı değişkenleri ekleyerek veya kaldırarak farklı model özelliklerini deneyin.

```python
new_df = df[["FullBath","MSSubClass","LotFrontage","MoSold","SalePrice"]].dropna()
X = new_df[["FullBath","MSSubClass","LotFrontage","MoSold"]]
Y = new_df["SalePrice"].dropna()

lrm = LinearRegression()
lrm.fit(X, Y)

x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```
